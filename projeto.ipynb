```python
import numpy as np         
import pandas as pd         


np.random.seed(42)

# Quantidade de registros (linhas)
n = 200
tempo = range(n)

# Gerando dados de temperatura:
temperatura = np.random.normal(75, 2, n)

# Gerando dados de vibração:
vibracao = np.random.normal(5, 0.5, n)

# Gerando dados de uso da máquina:
uso = np.random.normal(70, 10, n)

#  Inserindo anomalias manualmente (simulando falhas)

# Temperaturas muito altas (anomalias)
temperatura[50] = 95
temperatura[120] = 92

# Vibração fora do normal
vibracao[80] = 9
vibracao[150] = 8.5

# Uso da máquina muito alto
uso[30] = 95
uso[160] = 98

# Criando um DataFrame (tabela organizada)
df = pd.DataFrame({
    "tempo": tempo,
    "temperatura": temperatura,
    "vibracao": vibracao,
    "uso_maquina": uso
})


df.head(200)
```

```python
# Calculando média e desvio padrão de cada variável
media_temp = df["temperatura"].mean()
desvio_temp = df["temperatura"].std()

media_vib = df["vibracao"].mean()
desvio_vib = df["vibracao"].std()

media_uso = df["uso_maquina"].mean()
desvio_uso = df["uso_maquina"].std()

# Calculo de Z-score para cada coluna
df["z_temp"] = (df["temperatura"] - media_temp) / desvio_temp
df["z_vib"] = (df["vibracao"] - media_vib) / desvio_vib
df["z_uso"] = (df["uso_maquina"] - media_uso) / desvio_uso

#  Criando lógica de alerta 
def classificar_linha(row):
  # Aqui pego maior desvio
    maior_desvio = max(abs(row["z_temp"]), abs(row["z_vib"]), abs(row["z_uso"]))
    
    # Classificação baseada em regras de desvio
    if maior_desvio > 3:
        return "Crítico"
    elif maior_desvio > 2:
        return "Atenção"
    else:
        return "Normal"


df["alerta"] = df.apply(classificar_linha, axis=1)

df.head(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))

# Linha da temperatura
plt.plot(df["tempo"], df["temperatura"], label="Temperatura")

# Destacar pontos críticos
criticos = df[df["alerta"] == "Crítico"]

plt.scatter(criticos["tempo"], criticos["temperatura"], label="Crítico")

plt.title("Monitoramento de Temperatura com Detecção de Anomalias")
plt.xlabel("Tempo")
plt.ylabel("Temperatura")
plt.legend()
plt.grid()

plt.show()
```
